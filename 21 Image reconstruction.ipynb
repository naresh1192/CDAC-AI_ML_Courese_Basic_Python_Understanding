```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img1=cv2.imread('images1/cat_damaged.png')
```

```python
cv2.imshow('img', img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
height, width = img1.shape[:2]
```

```python
mask = np.zeros([height, width ,3], dtype = np.uint8)
```

```python
cv2.imshow('img', mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
for i in range(height):
    for j in range(width):
        if img1[i,j].sum()==0:
            mask[i,j]=[255,255,255]
cv2.imshow('img', mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
#mask = cv2.cvtColor(mask, cv2.COLOR_RGB2BGR)
```

```python
dst = cv2.inpaint(img1, mask, 3, cv2.INPAINT_NS)
```

```python
cv2.imshow('img', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,3,1)
plt.title('Original Image')
plt.imshow(img1)

plt.subplot(1,3,2)
plt.title('error detection img')
plt.imshow(mask)

plt.subplot(1,3,3)
plt.title(' reconsted image')
plt.imshow(dst)
```
