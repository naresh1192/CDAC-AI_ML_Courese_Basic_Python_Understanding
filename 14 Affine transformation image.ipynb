```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('images1/Sachin.jpg')
img.size
img.shape
```

```python
tx = 50
ty = 100
angle = 45
scale = 1.2
shear_x = 0.3
shear_y = 0.2
```

```python
M = np.float32([[scale*np.cos(np.radians(angle)), shear_x, tx], 
                [shear_y, scale*np.sin(np.radians(angle)), ty]])
```

```python
img_affine = cv2.warpAffine(img, M,(1075, 850) )
cv2.imshow('Affine Trns img', img_affine)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.subplot(1,2,1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
img_affine = cv2.cvtColor(img_affine, cv2.COLOR_BGR2RGB)
plt.subplot(1,2,2)
plt.imshow(img_affine)
```

```python
final_img = np.vstack([img, img_affine])
final_img = cv2.cvtColor(final_img, cv2.COLOR_BGR2RGB)
plt.imshow(final_img)
```
