```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
num = [np.random.randint(0,100) for x in range(1000)] #1 to 100, 1000 values are generated
```

```python
print(num)
```

```python
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Range')
plt.xticks(range(10,101,10))
plt.grid()
plt.hist(num, bins = range(10,101,10));
```
