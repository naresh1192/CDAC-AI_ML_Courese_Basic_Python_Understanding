```python
#kernel avg
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('images1/lenna.png')
```

```python
kernel = np.ones((3,3), np.float32)/9
```

```python
smoothed_img = cv2.filter2D(img, -1, kernel)
```

```python
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('img2', smoothed_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,2,1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
smoothed_img = cv2.cvtColor(smoothed_img, cv2.COLOR_BGR2RGB)
plt.title('Smoothened Image')
plt.imshow(smoothed_img)
```

```python
#kernel: gaussian distribution blurr
img2 = cv2.imread('images1/lenna.png')
kernel_size = 5
kernel = cv2.getGaussianKernel(kernel_size, 1.25) #1.25 is gaussian random number
blurred_img2 = cv2.filter2D(img2, -1, kernel)
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,2,1)
#img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.title('Original Image')
plt.imshow(img2, 'gray')

plt.subplot(1,2,2)
#smoothed_img = cv2.cvtColor(smoothed_img, cv2.COLOR_BGR2RGB)
plt.title('Blurred Image')
plt.imshow(blurred_img2, 'gray')
plt.show()
```
