```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
#Sobel Detection of edge
img=cv2.imread('images1/lenna.png',0)
```

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = 3)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize = 3)
combined_sobel = cv2.add(sobelx, sobely)
```

```python
final_img = np.hstack([sobelx, sobely, combined_sobel])
```

```python
cv2.imshow('All images', final_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,5,1)
plt.title('Original Image')
plt.imshow(img, cmap = 'gray')

plt.subplot(1,5,2)
plt.title(' Sobelx Image')
plt.imshow(abs(sobelx), cmap = 'gray')

plt.subplot(1,5,3)
plt.title('Sobely Image')
plt.imshow(abs(sobely), cmap = 'gray')

plt.subplot(1,5,4)
plt.title('Combined Image')
plt.imshow(abs(combined_sobel), cmap = 'gray')

plt.subplot(1,5,5)
plt.title('Stack Sobel')
plt.imshow(abs(final_img), cmap = 'gray')
```

```python
#Canny edges detection method

img=cv2.imread('images1/lenna.png',0)
edges = cv2.Canny(img, 50, 150)
cv2.imshow('img', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img, cmap = 'gray')

plt.subplot(1,2,2)
plt.title(' Canny Edge Image')
plt.imshow(edges, cmap = 'gray')
```
