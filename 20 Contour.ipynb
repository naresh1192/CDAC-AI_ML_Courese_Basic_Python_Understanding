```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img=cv2.imread('images1/circle.png',0)
thresh = cv2.threshold(img, 127,255, cv2.THRESH_BINARY)[1]
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
drawC = cv2.drawContours(img, contours, -1, (0,255,0),2)

cv2.imshow('img', thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('img', drawC)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,3,1)
plt.title('Original Image')
plt.imshow(img, cmap = 'gray')

plt.subplot(1,3,2)
plt.title('Threshold Image')
plt.imshow(thresh, cmap = 'gray')

plt.subplot(1,3,3)
plt.title(' Contour Edge Detection')
plt.imshow(drawC, cmap = 'gray')
```
