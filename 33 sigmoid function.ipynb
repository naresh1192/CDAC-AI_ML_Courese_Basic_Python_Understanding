```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.arange(-10, 10, 0.25)
print('The array of x:\n', x)
```

```python
y = 1 / (1 + np.exp(-x)) #sigmoid fun is always in between 0 and 1
print('The array of y:\n',y)
```

```python
plt.plot(x,y)
plt.grid()
```
