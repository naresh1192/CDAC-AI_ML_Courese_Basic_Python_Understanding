```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1=cv2.imread('images1/lenna.png')
img2=cv2.imread('images1/sun.png')
cv2.imshow('img1', img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('img2', img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1.shape, img2.shape
```

```python
added_img = cv2.add(img1, img2)
cv2.imshow('img', added_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
added_img = cv2.cvtColor(added_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (16,9))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.imshow(added_img)
```

```python
subst_img = cv2.substract(img1, img2)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
added_img = cv2.cvtColor(added_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (16,9))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.imshow(subst_img)
```

```python
#blending of image
alpha=0.75
blended_image = cv2.addWeighted(img1,alpha, img2, 1-alpha, 0)
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.imshow(blended_image)
```

```python
bright_img = cv2.multiply(img1, 1.5)
plt.figure(figsize = (16,9))
plt.subplot(1,2,1)
plt.imshow(img1)
plt.subplot(1,2,2)
plt.imshow(bright_img)
```
