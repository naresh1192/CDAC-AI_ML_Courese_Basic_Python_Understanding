```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img1=cv2.imread('images1/input1.png')
img2=cv2.imread('images1/input2.png')
cv2.imshow('img1', img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('img2', img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
dest_and = cv2.bitwise_and(img1, img2, mask = None)
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.imshow(dest_and)
```

```python
dest_or = cv2.bitwise_or(img1, img2, mask = None)
plt.figure(figsize = (16,9))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.imshow(dest_or)
```

```python
dest_xor = cv2.bitwise_xor(img1, img2, mask = None)
plt.figure(figsize = (16,9))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.imshow(dest_xor)
```

```python
dest_not = cv2.bitwise_not(img1, mask = None)
plt.figure(figsize = (16,9))
plt.subplot(1,2,1)
plt.imshow(img1)
plt.subplot(1,2,2)
plt.imshow(dest_not)
```

```python
dest_not = cv2.bitwise_not(img2, mask = None)
plt.figure(figsize = (16,9))
plt.subplot(1,2,1)
plt.imshow(img2)
plt.subplot(1,2,2)
plt.imshow(dest_not)
```
