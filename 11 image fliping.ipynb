```python
# fliping an image
import cv2
# load an image
image = cv2.imread('images/img1.jpg')
flipped_horizontal = cv2.flip(image,0)
flipped_vertical = cv2.flip(image,1)
flipped_diagonal = cv2.flip(image,-1)

image = cv2.imread('images/img1.jpg')
flipped_hor = cv2.cvtColor(flipped_horizontal, cv2.COLOR_BGR2RGB)
flipped_ver = cv2.cvtColor(flipped_vertical, cv2.COLOR_BGR2RGB)
flipped_diag = cv2.cvtColor(flipped_diagonal, cv2.COLOR_BGR2RGB)
original_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

cv2.imshow('Original Image', original_image)

cv2.imshow('Flipped image', flipped_hor)

cv2.imshow('Flipped image', flipped_ver)

cv2.imshow('Flipped image', flipped_diag)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
import matplotlib.pyplot as plt
plt.subplot(1,4,1)
plt.imshow(image)
plt.subplot(1,4,2)
plt.imshow(flipped_hor)
plt.subplot(1,4,3)
plt.imshow(flipped_ver)
plt.subplot(1,4,4)
plt.imshow(flipped_diag)
```
