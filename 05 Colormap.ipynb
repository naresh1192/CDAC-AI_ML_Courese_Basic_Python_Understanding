```python
#05 colormaps
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('images/img1.jpg', 0)
plt.imshow(img, cmap ='gray')
```

```python
#sequencial color map
#for gray
plt.figure(figsize = (16,9))
plt.subplot(2,5,1)
plt.title('Gray')
plt.imshow(img, cmap = 'gray')
#for blue
plt.subplot(2,5,2)
plt.title('Blues')
plt.imshow(img, cmap = 'Blues')
#for green
plt.subplot(2,5,3)
plt.title('Greens')
plt.imshow(img, cmap = 'Greens')
#for red
plt.subplot(2,5,4)
plt.title('Reds')
plt.imshow(img, cmap = 'Reds')
#for yellow
plt.subplot(2,5,5)
plt.title('Purples')
plt.imshow(img, cmap = 'Purples')
#for oranges
plt.subplot(2,5,6)
plt.title('BuPu')
plt.imshow(img, cmap = 'BuPu')
#for purple red PuRd
plt.subplot(2,5,7)
plt.title('PuRd')
plt.imshow(img, cmap = 'PuRd')
#for bluegreen BuGn
plt.subplot(2,5,8)
plt.title('BuGn')
plt.imshow(img, cmap = 'BuGn')
#for yellowbgreenblue YlGnBu
plt.subplot(2,5,9)
plt.title('YlGnBu')
plt.imshow(img, cmap = 'YlGnBu')
#for yellowblue YlBu
plt.subplot(2,5,10)
plt.title('YlGn')
plt.imshow(img, cmap = 'YlGn')
```

```python
#Divergent color map
#for gray
plt.figure(figsize = (16,9))
plt.subplot(2,5,1)
plt.title('seismic')
plt.imshow(img, cmap = 'seismic')
#for blue
plt.subplot(2,5,2)
plt.title('bwr')
plt.imshow(img, cmap = 'bwr')
#for green
plt.subplot(2,5,3)
plt.title('coolwarm')
plt.imshow(img, cmap = 'coolwarm')
#for red
plt.subplot(2,5,4)
plt.title('Spectral')
plt.imshow(img, cmap = 'Spectral')
#for yellow
plt.subplot(2,5,5)
plt.title('RdBu')
plt.imshow(img, cmap = 'RdBu')
#for oranges
plt.subplot(2,5,6)
plt.title('RdGy')
plt.imshow(img, cmap = 'RdGy')
#for purple red PuRd
plt.subplot(2,5,7)
plt.title('PuOr')
plt.imshow(img, cmap = 'PuOr')
#for bluegreen BuGn
plt.subplot(2,5,8)
plt.title('BrBG')
plt.imshow(img, cmap = 'BrBG')
#for yellowbgreenblue YlGnBu
plt.subplot(2,5,9)
plt.title('PRGn')
plt.imshow(img, cmap = 'PRGn')
#for yellowblue YlBu
plt.subplot(2,5,10)
plt.title('PiYG')
plt.imshow(img, cmap = 'PiYG')
```

```python
#Qualitative color map

plt.figure(figsize = (16,9))
plt.subplot(2,5,1)
plt.title('Set1')
plt.imshow(img, cmap = 'Set1')

plt.subplot(2,5,2)
plt.title('Set2')
plt.imshow(img, cmap = 'Set2')

plt.subplot(2,5,3)
plt.title('Set3')
plt.imshow(img, cmap = 'Set3')

plt.subplot(2,5,4)
plt.title('tab10')
plt.imshow(img, cmap = 'tab10')

plt.subplot(2,5,5)
plt.title('tab20')
plt.imshow(img, cmap = 'tab20')

plt.subplot(2,5,6)
plt.title('tab20b')
plt.imshow(img, cmap = 'tab20b')

plt.subplot(2,5,7)
plt.title('tab20c')
plt.imshow(img, cmap = 'tab20c')

plt.subplot(2,5,8)
plt.title('Dark2')
plt.imshow(img, cmap = 'Dark2')

plt.subplot(2,5,9)
plt.title('Accent')
plt.imshow(img, cmap = 'Accent')

plt.subplot(2,5,10)
plt.title('Paired')
plt.imshow(img, cmap = 'Paired')
```

```python
#cyclic color map

plt.figure(figsize = (16,9))
plt.subplot(1,3,1)
plt.title('hsv')
plt.imshow(img, cmap = 'hsv')

plt.subplot(1,3,2)
plt.title('twilight')
plt.imshow(img, cmap = 'twilight')

plt.subplot(1,3,3)
plt.title('twilight_shifted')
plt.imshow(img, cmap = 'twilight_shifted')
```
