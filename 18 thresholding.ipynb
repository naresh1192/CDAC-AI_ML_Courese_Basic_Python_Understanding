```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img=cv2.imread('images1/lenna.png',0)
```

```python
new_img = img > 127
inv_img = img < 127
```

```python
new_img, inv_img
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,3,1)
plt.title('Original Image')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img, cmap = 'gray')

plt.subplot(1,3,2)
plt.title('Binary Image')
plt.imshow(new_img, cmap = 'gray')

plt.subplot(1,3,3)
plt.title('Inv Binary Image')
plt.imshow(inv_img, cmap = 'gray')
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,3,1)
plt.title('Original Image')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)

plt.subplot(1,3,2)
plt.title('Binary Image')
plt.imshow(new_img, cmap = 'gray')

plt.subplot(1,3,3)
plt.title('Inv Binary Image')
plt.imshow(inv_img, cmap = 'gray')
```

```python
#otsu thresholding

thresh_val, thresh_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)
```

```python
thresh_val
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,4,1)
plt.title('Original Image')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)

plt.subplot(1,4,2)
plt.title('Binary Image')
plt.imshow(new_img, cmap = 'gray')

plt.subplot(1,4,3)
plt.title('Inv Binary Image')
plt.imshow(inv_img, cmap = 'gray')

plt.subplot(1,4,4)
plt.title('Otsu threshold Image')
plt.imshow(thresh_otsu, cmap = 'gray')
```
