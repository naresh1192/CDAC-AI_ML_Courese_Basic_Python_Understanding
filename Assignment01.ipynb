```python
width = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))
frames = int(vid.get(cv2.CAP_PROP_FRAME_COUNT))
fps = int(vid.get(cv2.CAP_PROP_FPS))
print('Video Resolution', height, width)
print('Total frames', frames)
print('FPS', fps)
```

```python
vid = cv2.VideoCapture('images1/nature.mp4')
while True:
   ret, frame = vid.read()
   if not ret:
       break
# Processing the frame
   cv2.imshow('Video', frame)
   if cv2.waitKey(3) == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
vid = cv2.VideoCapture('images1/nature.mp4')
# get the first frame and see if successful
success, frame = vid.read()
# initiating the count
count = 1
# keep extracting the frames as long as success is true
while success:
      cv2.imwrite(f'video_frames_nature/{count}.jpg', frame)
      success, frame = vid.read()
      count += 1
```

```python
timestamp = '00:00:50'
timestamp.split(':')
#get frame at specific timestamp
hours, minutes, seconds = [float(i) for i in timestamp.split(':')]
print(hours, minutes, seconds)
```

```python
trigger_frame = hours*3600*fps + minutes*60*fps + seconds*fps
print(frames, trigger_frame)
```

```python
# Obtaining selected frame by timestamp
vid.set(1, trigger_frame)
success, frame = vid.read()
# writing the image 
if success:
    cv2.imwrite('nature1.jpg', frame)
```
