```python
#121Matching two images
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img1=cv2.imread('images1/eiphel1.png',0)
img2=cv2.imread('images1/eiphel2.png',0)
```

```python
plt.figure(figsize = (16,9))
plt.subplot(1,2,1)
plt.title('Image1')
plt.imshow(img1, cmap = 'gray')

plt.subplot(1,2,2)
plt.title('Image2')
plt.imshow(img2, cmap = 'gray')
```

```python
sift = cv2.SIFT_create()
```

```python
keypoitns_1, descriptors_1 = sift.detectAndCompute(img1,None)
keypoitns_2, descriptors_2 = sift.detectAndCompute(img2,None)
len(keypoitns_1), len(keypoitns_2)
```

```python
bf = cv2.BFMatcher(cv2.NORM_L1)
matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)
img3 = cv2.drawMatches(img1,keypoitns_1, img2,keypoitns_2, matches[0:40], img2, flags = 2)
```

```python
plt.imshow(img3)
```
