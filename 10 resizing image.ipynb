```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
# load an image
image = cv2.imread('images/img1.jpg')

# resize an image
image_resize = cv2.resize(image,  (1000, 2000))
# show an resized image
cv2.imshow('image_resize',image_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(image_resize)
plt.figure(figsize= (16,9))
```

```python
import cv2
# load an image
image = cv2.imread('images/img1.jpg')

# resize an image
image_resize = cv2.resize(image, None, fx=1.75, fy=1.75)
# show an resized image
cv2.imshow('image_resize',image_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(image_resize)
plt.figure(figsize= (16,9))
```

```python
import matplotlib.pyplot as plt
from skimage.transform import resize
import numpy as np
# load an image
image = plt.imread('images/img1.jpg')

# resize an image
image_resize = resize(image, (1000, 2000), anti_aliasing = True)

# show an resized image
plt.subplot(1,2,1)
plt.imshow(image_resize)
plt.title('Resize Image')

plt.subplot(1,2,2)
plt.imshow(image)
plt.title('Original Image')
```
