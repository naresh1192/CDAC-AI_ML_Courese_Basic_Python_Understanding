```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('images1/Sachin.jpg')
```

```python
shear_x = 0.25
shear_y = 0.25
```

```python
M = np.float32([[1, shear_x, 0],[shear_y, 1, 0],[0, 0, 1]]) 
```

```python
rows, cols, _ = img.shape
T = np.float32([[1, 0, 0],[0, 1, 50]]) 
```

```python
img_trans = cv2.warpAffine(img, T, (int(cols*1.5), int(rows*1.5)))
```

```python
sheard_img = cv2.warpPerspective(img_trans, M, (int(cols*1.5), int(rows*1.5)) )
```

```python
cv2.imshow('img', sheard_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
sheard_img = cv2.cvtColor(sheard_img, cv2.COLOR_BGR2RGB)
plt.imshow(sheard_img)
```
